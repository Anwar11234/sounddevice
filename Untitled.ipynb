```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
from scipy.fftpack import fft
import math

t = np.linspace(0,3,12*1024)

Lfi = np.array([174.61, 130.81, 146.83, 164.81,174.61, 196, 220, 246.93])
Rfi= np.array([174.61,493.88,440,392,343.23,232.63,293.66,261.63])

tL= np.array([0, 0.4, 0.8, 1.3, 1.9, 2.3, 2.5, 2.7])
TR = np.array([0.3, 0.3, 0.4, 0.5, 0.3, 0.1, 0.1, 0.2])

N=7
h=0
d=0
while (d<N):
    Fi = Lfi[d]
    fi = Rfi[d]
    ti = tL[d]
    Ti = TR[d]
    noteL = np.sin(2*np.pi*Fi*t)
    noteR = np.sin(2*np.pi*fi*t)
    h = h + (noteL+noteR)*((t>=ti)&(t<=(ti+Ti)))
    d = d+1
```

```python
plt.plot(t,h)
```

```python
sd.play(h,3*1024)
```

```python
t = np.linspace(0,3,12*1024)

Lfi = np.array([174.61, 130.81, 146.83, 164.81,174.61, 196, 220, 246.93])
Rfi= np.array([174.61,493.88,440,392,343.23,232.63,293.66,261.63])

tL= np.array([0, 0.4, 0.8, 1.3, 1.9, 2.3, 2.5, 2.7])
TR = np.array([0.3, 0.3, 0.4, 0.5, 0.3, 0.1, 0.1, 0.2])
N=8
c=0
x=0
while (c<N):
    Fi = Lfi[c]
    fi = Rfi[c]
    ti = tL[c]
    Ti = TR[c]
    nL = np.sin(2*np.pi*Fi*t)
    nR = np.sin(2*np.pi*fi*t)
    x = x + (nL+nR)*((t>=ti)&(t<=(ti+Ti)))
    c = c+1

N = 3*1024
f = np. linspace(0 , 512 , int(N/2))

x_f = fft(x)
x_f = 2/N * np.abs(x_f [0:np.int(N/2)])

fn1 , fn2 = (np. random. randint(0, 512, 2))
n = (np.sin(2*t*np.pi*fn1)+np.sin(2*t*np.pi*fn2))

xn = x+n
xnf = fft(xn)
xnf = (2/N * np.abs(xnf [0:np.int(N/2)]))

z = (np.where(xnf>math.ceil(np.max(x))))
i1 = z[0][0]
i2 = z[0][1]

f1 = int(f[i1])
f2 = int(f[i2])

xf = (xn - (np.sin(2*t*np.pi*f1)+np.sin(2*t*np.pi*f2)))

sd.play(xf, 3*1024)

xff = fft(xf)
xff = 2/N * np.abs(xff [0:np.int(N/2)])
```

```python
plt.figure()
plt.subplot(3,1,1)
plt.plot(t,x)
plt.subplot(3,1,2)
plt.plot(t,xn)
plt.subplot(3,1,3)
plt.plot(t,xf)
```
